```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('data_clean.csv')
df.head(10)
```

```python
df.info();
```

```python
cereal_corr = df.corr()
cereal_corr
```

```python
fig, ax = plt.subplots(figsize=(12,9))

sns.heatmap(data = cereal_corr, annot=True, fmt='.2f', cmap='Blues',
           vmin=-1, vmax=1, linecolor='white', linewidths=0.5);
```

```python
plt.figure(figsize=(6,9))

plt.scatter(df['Kinh độ'], df['Vĩ độ'], c=np.log10(df['Mật độ dân số']));
plt.title('Phân bố bản ghi theo kinh độ và vĩ độ')
plt.xlabel('Kinh độ')
plt.ylabel('Vĩ độ')
plt.colorbar(label='log$_{10}$(df[Mật độ dân số])');
```

```python
rng = np.random.RandomState(0)

x = df['Diện tích đất']
y = df['Giá']
colors = rng.rand(79026)
sizes = pd.DataFrame(1, index=np.arange(1, 79027), columns=np.arange(1)).values.tolist()

fig, ax = plt.subplots()
ax.set(title='Phân bố Giá nhà theo Diện tích đất',
      xlabel='Diện tích đất (m2)',
      ylabel='Giá (triệu)');
ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.3);
```
